```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('coffeedata.csv')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
missing_values_sum = df.isnull().sum()
print(missing_values_sum)
missing_values_percent = df.isnull().mean()*100
print(missing_values_percent)
```

```python
# Replace nulls with 'unknown'
df.roast_level.fillna('unknown', inplace=True)
df.est_price.fillna('unknown', inplace=True)
df.bottom_line.fillna('unknown', inplace=True)

df.dropna()
```

```python
missing_values_sum = df.isnull().sum()
print(missing_values_sum)
```
